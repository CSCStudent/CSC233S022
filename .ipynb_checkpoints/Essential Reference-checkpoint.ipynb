## Variables and Arithmetic Expressions

Python is a dynamically typed language where variable names are bound to different values, possibly of different
types, during program execution.  The assignment operator simply creates an association between the variable name and its value.  The following listing shows the use of variables and expressions.

```python
principal = 1000
rate = 0.05
numyears = 5
year = 1
while year <= numyears:
    principal = principal * (1 + rate)
    print ("year: %d principal %.2f" %(year, principal))
    year += 1
```

## Lists

Lists are a sequence of arbitrary objects that you can create by enclosing the values in square brackets.  The list is managed very much like a dynamic array where the first element has an index of 0.

```python
names = []     #an empty list
names = list() #an empty list
names = ["Dave", "Mark", "Phil"]    #static definition
print(names)
names[1] = "Bob"  #dynamic definition
print(names)
```

The objects in a list can be of varying types:

```python
things = ["Jack", 123, True, ["A","B","C"]]
print (things)
```

## Tuples

To create a simple data structure, you can pack a collection of values together into a single object using a <b>tuple</b>.

```python
stock = ("GOOG", 100, 490.10)
print(stock)

networkAddress = ("www.python.org", 80)
print("host: %s port: %d" %(networkAddress))

person = ("Jack", "Sparrow", "970-123-4567")
fn, ln, phone = person
print("{} {} {}".format(fn, ln, phone))
```

```python
portfolio = []
filename = "portfolio.txt"
for line in open(filename):
    fields = line.split(",")
    name = fields[0]
    shares = int(fields[1])
    price = float(fields[2])
    stock = (name, shares, price)
    portfolio.append(stock)

for stk in portfolio:
    print("name: %s shares: %d price: %.2f" %(stk))
```

## Sets

A <b>set</b> is used to contain an unordered collection of objects where there is no duplicate.  To create a set, use the constructor <b>set()</b> passing in a list of values.

```python
s = set([1,3,5,7,9, 3])
print(s)
```

```python
t = set("Hello")
print(t)

t.add('x')           #add a single element
t.update([2,4,6,8])  #add multiple elements

print(t)
```

## Dictionaries

A <b>dictionary</b> is an associative array or hash table of <key, value> pairs that contains objects indexed by the keys.  You create dictionaries by enclosing the key/values in curly braces "{ }".  Dictionaries can be used as a container for performing fast lookups on unordered data.

Dictionaries are probably the most finely tuned data structure in Python.  So if you're merely trying to store and work with data it's better to use a dictionary than any other type object.

```python
#defining the dictionary
stock = {}      #defining an empty dictionary
stock = dict()  #using the constructor

stock = { "GOOG" : 490.10,
         "APPL" : 123.50,
         "HPE" : 15.89
}
print(stock)

stock["APPL"] = 128.33

print(stock.get("APPL"))

if "GOOG" in stock:
    print("value of GOOG: {}".format(stock.get("GOOG")))
    
syms = list(stock)    #obtain a list of keys
print(syms)
    
del stock["HPE"]  #delete and element from dictionary
print(stock)
```

## Iterations and Looping

Python loop construct is comprised of the <b>for</b> and the <b>while</b> semantics.

```python
#using the for loop to iterate over items
for n in [1,2,3,4,5,6]:
    print("2 to the %d power is %d" %(n, 2**n))
```

```python
##
#  This program computes the time required to double an investment.
#

# Create constant variables.
RATE = 5.0
INITIAL_BALANCE = 10000.0
TARGET = 2 * INITIAL_BALANCE
      
# Initialize variables used with the loop.
balance = INITIAL_BALANCE
year = 0

# Count the years required for the investment to double.
while balance < TARGET :
   year = year + 1
   interest = balance * RATE / 100
   balance = balance + interest
   
# Print the results.   
print("The investment doubled after", year, "years.")

```
