# ToolBox 2.1 Symbolic Processing with SymPy

The SymPy package can be used for symbolic mathematics.  Python can compute the value of mathematical expressions, plot a function, and compute a variety of formulas.

## Getting Started

Before being able to use a package such as SymPy, it must be installed on your system.
The functionality of a third-party code package is organized in one or more modules.  You need to import the modules that you need.

```python
from sympy import *
```

## Working with expressions

One useful function is sympify() takes a string argument and turns it into a SymPy expression object.

```python
x = sympify("x * x**2")
print(x)
```

Alternately, you can first define the symbolic expression object x and store it in a variable (it is covenient to name the variable by the same name).  The use the operators and functions to build upa SymPy expression.

```python
f = sympify("x * x**2")
print(f)
```

```python
x = sympify("x")
f = log(x)
plot(f)
```

Working with symbol expression objects is useful for simplifying algebraic expressions:

```python
print( expand("(x-1) * (x+1)"))
print( expand("(x-1) ** 5"))
```

You may want to plot a function using the plot().  Here we designate of range of -20 to 20 for x.

```python
x = sympify("x")
f = x ** 2 * sin(x)
print(f)
plot(f, (x,-20,20))
```

Sympy can compute the derivative and integral:

```python
print("Derivative of {} is {}" .format(f, diff(f)))
```

```python
g = integrate(f)
print("Integral g(x) is {}", g)
print("g(0) is {0:10.2f}".format(g.subs(x,0).evalf()))
```
