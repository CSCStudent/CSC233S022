# Plotting Simple Graphs

## Creating a Graph

To create and display a graph using the matplotlib module, carry out the following steps:

```python
from matplotlib import pyplot
```

The pyplot module has functions for adding elements to and displaying a graph.
To plot data on the graph, you invoke function in the pyplot module.  Here, we use
the bar() function to plot severval values as bars on a graph.

```python
pyplot.bar(1, 1.1)
pyplot.bar(2, 10.0)
pyplot.bar(3, 35.4)
pyplot.bar(4, 44.5)
pyplot.bar(5, 61.0)
```

You can also use a list:

```python
pyplot.bar([1,2,3,4,5],[1.1, 10.0, 35.4, 44.5, 61.0])
```

Improve the appearance of the graph:

```python
pyplot.xlabel("Month")
pyplot.ylabel("Temp")
```

After you've configured the graph, call the show() function:

```python
pyplot.bar([1,2,3,4,5],[1.1, 10.0, 35.4, 44.5, 61.0])
pyplot.xlabel("Month")
pyplot.ylabel("Temp")
pyplot.show()
```

You can change the color of a bar in a bar char by calling: pyplot.bar(x,y,color="..." where valid color strings are listed in the below.

<p>blue - 'b'</p>
<p>green - 'g'</p>
<p>red - 'r'</p>



```python
pyplot.bar([1,2,3],[1.1, 10.0, 35.4], color='r')
pyplot.bar([4,5],[44.5, 61.0], color ='g')
pyplot.show()
```

# Creating a Line Graph

A line graph connects data points with line sements.  You can pass lists of
x- and y- coordinates to the plot() function.

```python
pyplot.plot([1,2,3,4,5],[1.1, 10.0, 35.4, 44.5, 61.0])
pyplot.xlabel("Month")
pyplot.ylabel("Temp")
pyplot.show()
```
